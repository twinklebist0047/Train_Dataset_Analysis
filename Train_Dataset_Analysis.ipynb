## Detecting Outliers

1. if the data is following normal distribution, anything beyond 3SD - Mean + 3SD car be considered as an outlier.
2. if the data does not follow normal distribution, using boxplot we can eliminate points beyond Q1 - 1.5 IQR and Q3 + 1.5 IQR

```python
Q1= np.percentile(train['Fare'],25)
Q3= np.percentile(train['Fare'],75)

outlier_low=Q1 - 1.5 * (Q3 - Q1)
outlier_high=Q3 + 1.5 * (Q3 - Q1)

train=train[(train['Fare']>outlier_low) & (train['Fare']<outlier_high)]
```
